```python
import cirq
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
'''Initialize named qutrits and create a circuit to measure them. Print the circuit and results.'''

i = cirq.NamedQid('i', dimension=3)
j = cirq.NamedQid('j', dimension=3)
k = cirq.NamedQid('k', dimension=3)
a = cirq.NamedQid('a', dimension=3)
b = cirq.NamedQid('b', dimension=3)
c = cirq.NamedQid('c', dimension=3)
d = cirq.NamedQid('d', dimension=3)
e = cirq.NamedQid('e', dimension=3)
circuit = cirq.Circuit(cirq.measure(i,j,k))
simulator = cirq.Simulator()
result1 = simulator.run(circuit, repetitions=10)

circuitReset = cirq.Circuit(cirq.reset(i), cirq.reset(j), cirq.reset(k))

print('Circuit:')
print(circuit)
print('Results:')
print(result1)
```

```python
'''
OFF STATE: Reset ijk to |0>. Print circuit and return 000 state. 
'''
def off_state(i, j, k):
    
    circuit_off = cirq.Circuit(circuitReset, cirq.measure(i, j, k))
    sim = cirq.Simulator()
    result = sim.run(circuit_off, repetitions=1)
    measurement = "".join((map(str,result.measurements.values())))
    vector = ''.join(str(v[0]) for v in result.measurements.values())
    print("Circuit off_state:"'\n', (circuit_off))
    print("Result off_state:"'\n', (result))
    print("Measurement vector:", vector)
    return(vector)
off_state(i,j,k)
```

```python
'''
ON/SCAN STATE: Reset ijk to |0>. X flip i to |1>. Print circuit and return 100 state. 
''' 
def on_state(i, j, k):
    circuit_on = cirq.Circuit(circuitReset, cirq.XPowGate(dimension=3, exponent=1).on(i), cirq.measure(i, j, k))
    sim = cirq.Simulator()
    result = sim.run(circuit_on, repetitions=1)
    measurement = "".join((map(str, result.measurements.values())))
    vector = ''.join(str(v[0]) for v in result.measurements.values())
    print("Circuit on_state:"'\n',(circuit_on))
    print("Result on_state:"'\n', (result))
    print("Measurement vector:", vector)
    return vector
on_state(i,j,k)
```

```python
'''
RECEIVE STATE: Reset ijk to |0>. X flip i and k. Print circuit and return 101 state. 
'''
def receive_state(i,j,k): 
    circuit_receive = cirq.Circuit(circuitReset, cirq.XPowGate(dimension=3, exponent=1).on(i), cirq.XPowGate(dimension=3, exponent=1).on(k), cirq.measure(i, j, k))
    sim = cirq.Simulator()
    result = sim.run(circuit_receive, repetitions=1)
    measurement = "".join((map(str,result.measurements.values())))
    vector = ''.join(str(v[0]) for v in result.measurements.values())
    print("Circuit receive_state:"'\n', (circuit_receive))
    print("Result"'\n', (result))
    print("Measurement vector:", vector)
    return(vector)
receive_state(i,j,k)
```

```python
'''
READ state: Reset ijk to |0>. X flip i and k. X flip j twice. Print circuit and return 121 state. 
'''
def read_state(i,j,k): 
    circuit_read = cirq.Circuit(circuitReset, cirq.XPowGate(dimension=3, exponent=1).on(i), cirq.XPowGate(dimension=3, exponent=1).on(k), cirq.XPowGate(dimension=3, exponent=2).on(j), cirq.measure(i,j,k)) 
    sim = cirq.Simulator()
    result = sim.run(circuit_read, repetitions=1)
    measurement = "".join((map(str,result.measurements.values())))
    vector = ''.join(str(v[0]) for v in result.measurements.values())
    print("Circuit read_state:"'\n', (circuit_read))
    print("Result read_state:"'\n', (result))
    print("Measurement vector:", vector)
    return(vector)
read_state(i,j,k)
```

```python
''' 
WRITE STATE: Reset ijk to |0>. X flip i and j. Print circuit and return 110 state. 
'''
def write_state(i,j,k):
    circuit_write = cirq.Circuit(circuitReset, cirq.XPowGate(dimension=3, exponent=1).on(i), cirq.XPowGate(dimension=3, exponent=1).on(j), cirq.measure(i, j, k))
    sim = cirq.Simulator()
    result = sim.run(circuit_write, repetitions=1)
    measurement = "".join((map(str,result.measurements.values())))
    vector = ''.join(str(v[0]) for v in result.measurements.values())
    print("Circuit write_state"'\n', (circuit_write))
    print("Result", (result))
    print("Measurement vector:", vector)
    return(vector)
write_state(i,j,k)
```

```python
'''
Send state: Reset ijk to |0>. X flip i and j. X flip k twice. Print circuit and return 112 state. 
'''

def send_state(i,j,k): #|010>
    circuit_send = cirq.Circuit(circuitReset, cirq.XPowGate(dimension=3, exponent=1).on(i), cirq.XPowGate(dimension=3, exponent=1).on(j), cirq.XPowGate(dimension=3, exponent=2).on(k), cirq.measure(i,j,k))
    sim = cirq.Simulator()
    result = sim.run(circuit_send, repetitions=1)
    measurement = "".join((map(str,result.measurements.values())))
    vector = ''.join(str(v[0]) for v in result.measurements.values())
    print("Circuit send_state:"'\n',(circuit_send))
    print("Result"'\n', print(result))
    print("Measurement vector:", vector)
    return(vector)
send_state(i,j,k)
```

```python
'''
Store state: Reset ijk to |0>. X flip i, j and k. Print circuit and return 111 state. 
'''

def store_state(i,j,k):
    circuit_store = cirq.Circuit(circuitReset, cirq.XPowGate(dimension=3, exponent=1).on(i), cirq.XPowGate(dimension=3, exponent=1).on(j).(i), cirq.XPowGate(dimension=3, exponent=1).on(k).(i), cirq.measure(i,j,k))
    sim = cirq.Simulator()
    result = sim.run(circuit_store, repetitions=1)
    measurement = "".join((map(str,result.measurements.values())))
    vector = ''.join(str(v[0]) for v in result.measurements.values())
    print("Circuit store_state:"'\n', (circuit_store))
    print("Result" '\n', print(result))
    print("Measurement vector:", vector)
    return(vector)
store_state(i,j,k)
```

```python
'''
Delete state: Reset ijk to |0>. X flip i. X flip j and k twice. X flip i to |1>. Print circuit and return 122 state. 
'''

def delete_state(i,j,k): # |100>
    circuit_delete = cirq.Circuit(circuitReset, cirq.XPowGate(dimension=3, exponent=1).on(i), cirq.XPowGate(dimension=3, exponent=2).on(j), cirq.XPowGate(dimension=3, exponent=2).on(k), cirq.measure(i,j,k))
    sim = cirq.Simulator()
    result = sim.run(circuit_delete, repetitions=1)
    measurement = "".join((map(str,result.measurements.values())))
    vector = ''.join(str(v[0]) for v in result.measurements.values())
    print("Circuit delete_state:"'\n', print(circuit_delete))
    print("Result" '\n', print(result))
    print("Measurement vector:", vector)
    return(vector)
delete_state(i,j,k)
```

These eight operations allow for smooth, logical transitions from individual states. 
OFF (000) --> ON/SCAN (100)
ON/SCAN (100) --> RECEIVE (101) OR WRITE (110)
RECEIVE (101) --> STORE (111) OR READ (121)
WRITE (110) --> STORE (111) OR SEND (112)
STORE (111) --> DELETE (122) OR READ (121) OR SEND (112)
